```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.array([
    [0, 2],
    [0, 4],
    [0, 0],
    [1, 2],
    [1, 4],
    [1, 0]
])
```

```python
max_iters = 100
k = 2
```

```python
def euclidean_distance(a, b):
  return np.sqrt(np.sum((a - b) ** 2))
```

```python
np.random.seed(42)
```

```python
centroid_idx = np.random.choice(len(X), k, replace=False)
centroids = X[centroid_idx].astype(float)
```

```python
centroids
```

```python

for _ in range(max_iters):
  labels = []
  # assigning centroids
  for x in X:
    distances = []
    for c in centroids:
      d = euclidean_distance(x, c)
      distances.append(d)
    closest_cluster_idx = np.argmin(distances)
    labels.append(closest_cluster_idx)

  labels = np.array(labels)

  # Updating Centroids
  new_centroids = []
  for i in range(k):
    cluster_points = X[labels == i]
    if len(cluster_points) > 0:
      new_center = np.mean(cluster_points, axis=0)
    else:
      new_center = centroids[i]
    new_centroids.append(new_center)

  new_centroids = np.array(new_centroids)

  # if converged, stop
  if np.allclose(new_centroids, centroids):
    break

  centroids = new_centroids
```

```python
print(f"Final centroids: {centroids}")
print(f"Final cluster labels: {labels}")
```
