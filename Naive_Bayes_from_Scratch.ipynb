```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/DiogoRibeiro7/c6590d0cf119e87c39e31c21a9c0f3a8/raw/4a8e3da267a0c1f0d650901d8295a5153bde8b21/PlayTennis.csv')
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
df.head()
```

```python
X.shape[0]
```

```python
def prior_probability(df, label):
  total_examples = X.shape[0]
  prior_probs = {}
  for label in np.unique(y):
    prior_probs[label] = np.sum(y == label) / total_examples
  return prior_probs
```

```python
priors = prior_probability(df, y)
```

```python
X.columns
```

```python
likelihoods = {}
for feature in X.columns:
  likelihoods[feature] = {}
  for value in np.unique(X[feature]):
    likelihoods[feature][value] = {}
    for label in np.unique(y):
      label_occurences = np.sum((X[feature] == value) & (y == label))
      total_label_occurences = np.sum(y == label)
      likelihoods[feature][value][label] = label_occurences / total_label_occurences
```

```python
likelihoods
```

```python
priors.items()
```

```python
for _, row in df.iterrows():
  for label in np.unique(y):
    for feature in X.columns:
      prob = likelihoods[feature][row[feature]][label]

```

```python
predictions = {}
for index, row in df.iterrows():
  for label in np.unique(y):
    predictions[label] = {}
    prob = 1
    for feature in X.columns:
      prob *= likelihoods[feature][row[feature]][label]
    predictions[label] = prob * priors[label]
```

```python
predictions
```

```python
def predict_row(row, priors):
  posteriors = {}

  for label in np.unique(y):
    prob = priors[label]
    predictions[label] = {}
    for feature in X.columns:
      prob *= likelihoods[feature][row[feature]][label]
    posteriors[label] = prob
  prediction = max(posteriors, key = posteriors.get)
  return prediction, posteriors
```

```python
preds=[]
for idx, row in df.iterrows():
  pred, _ = predict_row(row, priors)
  preds.append(pred)
  print(pred)
```

```python
def accuracy(y_pred, y_true):
  correct = 0
  for yt, yp in zip(y_true, y_pred):
    if yt == yp:
      correct += 1
  return correct / len(y)
```

```python
accuracy(preds, y)
```
